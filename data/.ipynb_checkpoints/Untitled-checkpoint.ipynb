```python
import fitz  # PyMuPDF
import json
from transformers import pipeline

from transformers import AutoTokenizer, AutoModelForSeq2SeqLM, pipeline

model_name = "google/flan-t5-base" 

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

generator = pipeline(
    "text-generation",  # works with this setup
    model=model,
    tokenizer=tokenizer,
    max_length=512
)

# -----------------------------
# STEP 1: Extract PDF text
# -----------------------------
def extract_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""

    for page in doc:
        text += page.get_text()

    return text


# -----------------------------
# STEP 2: Extract crop section
# -----------------------------
def extract_crop_section(text, crop_name, window=80):
    lines = text.split("\n")

    filtered = []

    for line in lines:
        if crop_name.lower() in line.lower():
            idx = lines.index(line)
            return "\n".join(lines[idx:idx+window])

    return None


# -----------------------------
# STEP 3: Convert to structured JSON
# -----------------------------
def convert_to_json(text_chunk, crop_name):
    prompt = f"""
You are an agriculture expert.

Extract ONLY structured information about the crop: {crop_name}.

Return STRICT JSON format:

{{
  "crop": "",
  "soil": "",
  "climate": "",
  "season": "",
  "stages": [],
  "fertilizer": {{"N":"", "P":"", "K":""}},
  "irrigation": ""
}}

If information is missing, write "Not available".

DO NOT explain anything.
DO NOT repeat input.
ONLY return JSON.

TEXT:
{text_chunk}
"""

    output = generator(prompt)[0]['generated_text']

    return output


# -----------------------------
# STEP 4: Save JSON
# -----------------------------
def save_json(data, filename):
    import json
    
    try:
        parsed = json.loads(data)
    except:
        print("⚠️ Invalid JSON, fixing...")
        parsed = {"error": data[:500]}  # partial save

    with open(filename, "w") as f:
        json.dump(parsed, f, indent=4)

# -----------------------------
# STEP 5: Full pipeline
# -----------------------------
def process_crop(pdf_path, crop_name):
    print(f"\n🔍 Processing: {crop_name}")

    text = extract_text(pdf_path)

    crop_text = extract_crop_section(text, crop_name)

    if not crop_text:
        print("❌ Crop not found")
        return

    json_data = convert_to_json(crop_text, crop_name)

    save_json(json_data, f"{crop_name}.json")


# -----------------------------
# RUN
# -----------------------------
if __name__ == "__main__":
    pdf_path = "books/crop_production_and _management.pdf"

    crops = ["rice"]  # start with one

    for crop in crops:
        process_crop(pdf_path, crop)
```

```python
import fitz  # PyMuPDF
import os
import json
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

# -----------------------------
# LOAD BETTER MODEL (Phi-3)
# -----------------------------
model_name = "microsoft/Phi-3-mini-4k-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype=torch.float32   # use float16 if GPU
)

model.eval()


# -----------------------------
# GENERATE FUNCTION
# -----------------------------
def generate(prompt):
    inputs = tokenizer(prompt, return_tensors="pt", truncation=True, max_length=2048)
    inputs = {k: v.to(model.device) for k, v in inputs.items()}

    outputs = model.generate(
        **inputs,
        max_new_tokens=300,
        do_sample=False
    )

    return tokenizer.decode(outputs[0], skip_special_tokens=True)


# -----------------------------
# FIND PDF AUTOMATICALLY
# -----------------------------
def get_pdf_path():
    for f in os.listdir("books"):
        if f.endswith(".pdf"):
            return os.path.join("books", f)
    raise FileNotFoundError("No PDF found in books folder")


# -----------------------------
# EXTRACT TEXT FROM PDF
# -----------------------------
def extract_text(pdf_path):
    doc = fitz.open(pdf_path)
    text = ""
    for page in doc:
        text += page.get_text()
    return text


# -----------------------------
# CLEAN TEXT (VERY IMPORTANT)
# -----------------------------
def clean_text(text):
    lines = text.split("\n")
    clean = []

    for line in lines:
        line = line.strip()

        if len(line) < 5:
            continue
        if line.lower().startswith("chapter"):
            continue
        if line.startswith("("):
            continue

        clean.append(line)

    return "\n".join(clean)


# -----------------------------
# EXTRACT CROP SECTION
# -----------------------------
def extract_crop_section(text, crop_name, window=50):
    text = clean_text(text)
    lines = text.split("\n")

    for i, line in enumerate(lines):
        if crop_name.lower() in line.lower():
            return "\n".join(lines[i:i+window])

    return None


# -----------------------------
# CONVERT TO STRUCTURED JSON
# -----------------------------
def convert_to_json(text_chunk, crop_name):
    prompt = f"""
You are an agriculture expert.

Extract structured data for crop: {crop_name}

Return STRICT JSON:
{{
  "crop": "",
  "soil": "",
  "climate": "",
  "season": "",
  "stages": [],
  "fertilizer": {{"N":"","P":"","K":""}},
  "irrigation": ""
}}

Rules:
- If missing → "Not available"
- Do NOT explain
- ONLY return JSON

TEXT:
{text_chunk[:1200]}
"""

    return generate(prompt)


# -----------------------------
# SAVE JSON SAFELY
# -----------------------------
def save_json(data, filename):
    try:
        parsed = json.loads(data)
    except:
        print("⚠️ JSON invalid → saving raw output")
        parsed = {"raw_output": data[:500]}

    with open(filename, "w") as f:
        json.dump(parsed, f, indent=4)

    print(f"✅ Saved: {filename}")


# -----------------------------
# MAIN PIPELINE
# -----------------------------
def process_crop(crop_name):
    print(f"\n🔍 Processing: {crop_name}")

    pdf_path = get_pdf_path()
    print("📄 Using:", pdf_path)

    text = extract_text(pdf_path)

    crop_text = extract_crop_section(text, crop_name)

    if not crop_text:
        print("❌ Crop not found in PDF")
        return

    result = convert_to_json(crop_text, crop_name)

    save_json(result, f"{crop_name}.json")


# -----------------------------
# RUN
# -----------------------------
if __name__ == "__main__":
    crops = ["paddy"]   # IMPORTANT (better than "rice")

    for crop in crops:
        process_crop(crop)
```

```python
import fitz
import os
import re
import json

BOOKS_PATH = "books"
OUTPUT_FILE = "clean_agri_dataset.json"

# -----------------------------
# KEYWORDS
# -----------------------------
FERTILIZERS = ["urea", "npk", "fertilizer", "manure", "super phosphate"]
PESTICIDES = ["spray", "insecticide", "chlorpyriphos", "monocrotophos", "carbaryl", "malathion"]
DISEASE_WORDS = ["disease", "fungus", "virus", "blight", "rust"]
CROPS = ["rice", "wheat", "maize", "cotton", "sugarcane"]

# -----------------------------
# EXTRACT TEXT
# -----------------------------
def extract_text(pdf):
    doc = fitz.open(pdf)
    text = ""
    for page in doc:
        text += page.get_text()
    return text.lower()

# -----------------------------
# CLEAN TEXT
# -----------------------------
def clean_text(text):
    text = re.sub(r'\n+', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text

# -----------------------------
# SPLIT
# -----------------------------
def split_sentences(text):
    return re.split(r'(?<=[.!?]) +', text)

# -----------------------------
# DETECT TYPE
# -----------------------------
def detect_type(sentence):
    if any(word in sentence for word in FERTILIZERS):
        return "fertilizer"
    if any(word in sentence for word in PESTICIDES):
        return "pesticide"
    if any(word in sentence for word in DISEASE_WORDS):
        return "disease"
    return None

# -----------------------------
# DETECT CROP
# -----------------------------
def detect_crop(sentence):
    for crop in CROPS:
        if crop in sentence:
            return crop
    return None

# -----------------------------
# EXTRACT CHEMICAL NAME
# -----------------------------
def extract_name(sentence):
    words = sentence.split()
    for w in words:
        if w in FERTILIZERS or w in PESTICIDES:
            return w
    return "unknown"

# -----------------------------
# FILTER BAD SENTENCES
# -----------------------------
def is_valid(sentence):
    if len(sentence) < 40:
        return False
    if "www." in sentence:
        return False
    if "lecture" in sentence:
        return False
    if "index" in sentence:
        return False
    return True

# -----------------------------
# BUILD DATASET
# -----------------------------
def build_dataset():
    data = []

    for file in os.listdir(BOOKS_PATH):
        if not file.endswith(".pdf"):
            continue

        print("Processing:", file)
        text = extract_text(os.path.join(BOOKS_PATH, file))
        text = clean_text(text)
        sentences = split_sentences(text)

        for sent in sentences:
            sent = sent.strip()

            if not is_valid(sent):
                continue

            crop = detect_crop(sent)
            dtype = detect_type(sent)

            if crop and dtype:
                entry = {
                    "crop": crop,
                    "type": dtype,
                    "name": extract_name(sent),
                    "content": sent
                }
                data.append(entry)

    return data

# -----------------------------
# SAVE
# -----------------------------
def save(data):
    with open(OUTPUT_FILE, "w") as f:
        json.dump(data, f, indent=2)
    print("Saved:", len(data))

# -----------------------------
# RUN
# -----------------------------
if __name__ == "__main__":
    dataset = build_dataset()
    save(dataset)
```

```python
import json
import re
from tqdm import tqdm

# -----------------------------
# CONFIG
# -----------------------------
INPUT_FILE = "clean_agri_dataset.json"     # your current file
OUTPUT_FILE = "agri2_clean_dataset.json"

# -----------------------------
# VALID KEYWORDS (expand anytime)
# -----------------------------
CHEMICAL_PATTERN = re.compile(
    r"(monocrotophos|carbaryl|chlorpyriphos|urea|dap|mancozeb|"
    r"dimethoate|malathion|zinc sulphate|acephate|fipronil|"
    r"spinosad|indoxacarb|thiamethoxam|azadirachtin|"
    r"carbendazim|streptomycin|copper oxychloride)",
    re.IGNORECASE
)

DISEASE_PATTERN = re.compile(
    r"(blast|blight|rust|mosaic|tungro|rot|smut|leaf curl|"
    r"dwarf|downy mildew|powdery mildew)",
    re.IGNORECASE
)

# -----------------------------
# CLEAN TEXT
# -----------------------------
def clean_text(text):
    text = text.lower()
    text = re.sub(r"\s+", " ", text)
    return text.strip()

# -----------------------------
# EXTRACT NAME
# -----------------------------
def extract_name(text, type_):
    if type_ in ["pesticide", "fertilizer"]:
        match = CHEMICAL_PATTERN.findall(text)
    else:
        match = DISEASE_PATTERN.findall(text)

    if match:
        return list(set(match))  # unique values
    return []

# -----------------------------
# VALIDATION FILTER
# -----------------------------
def is_valid(name, text):
    if not name:
        return False
    if len(text) < 40:
        return False
    if name in ["spray", "insecticide"]:
        return False
    return True

# -----------------------------
# PROCESS DATASET
# -----------------------------
def process_dataset(data):
    cleaned_data = []

    for item in tqdm(data):
        crop = item.get("crop", "").lower()
        type_ = item.get("type", "").lower()
        content = clean_text(item.get("content", ""))

        names = extract_name(content, type_)

        for name in names:
            if not is_valid(name, content):
                continue

            entry = {
                "crop": crop,
                "type": type_,
                "name": name,
                "description": content
            }

            cleaned_data.append(entry)

    return cleaned_data

# -----------------------------
# REMOVE DUPLICATES
# -----------------------------
def remove_duplicates(data):
    seen = set()
    unique_data = []

    for item in data:
        key = (item["crop"], item["type"], item["name"], item["description"])
        if key not in seen:
            seen.add(key)
            unique_data.append(item)

    return unique_data

# -----------------------------
# MAIN
# -----------------------------
def main():
    print("Loading dataset...")
    with open(INPUT_FILE, "r", encoding="utf-8") as f:
        data = json.load(f)

    print("Processing dataset...")
    cleaned = process_dataset(data)

    print("Removing duplicates...")
    cleaned = remove_duplicates(cleaned)

    print(f"Final dataset size: {len(cleaned)}")

    with open(OUTPUT_FILE, "w", encoding="utf-8") as f:
        json.dump(cleaned, f, indent=2)

    print(f"Saved to {OUTPUT_FILE}")


if __name__ == "__main__":
    main()
```
